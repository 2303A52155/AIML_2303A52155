BOOT STRAPING

```python
from sklearn.metrics import accuracy_score
from sklearn.utils import resample
from sklearn.svm import SVC
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier

def bootstrap(x_train, y_train, model, n_iterations=100):
  accuracy_scores = []
  for _ in range(n_iterations):
    x_resampled, y_resampled = resample(x_train, y_train, replace=True, random_state=None)
    model.fit(x_resampled, y_resampled)
    accuracy_scores.append(model.score(x_test, y_test))
  return accuracy_scores

bootstrap_scores = bootstrap(x_train, y_train, LogisticRegression())
print(f"Bootstrap Accuracy Scores (Logistic Regression): {bootstrap_scores}")

bootstrap_scores_perceptron = bootstrap(x_train,y_train,Perceptron())
print(f"Bootstrap Accuracy Scores (Perceptron): {bootstrap_scores_perceptron}")

bootstrap_scores_svm = bootstrap(x_train,y_train,SVC())
print(f"Bootstrap Accuracy Scores (SVM): {bootstrap_scores_svm}")

bootstrap_scores_knn = bootstrap(x_train, y_train, KNeighborsClassifier())
print(f"Bootstrap Accuracy Scores (KNN): {bootstrap_scores_knn}")

bootstrap_scores_ann = bootstrap(x_train, y_train, MLPClassifier(hidden_layer_sizes=(64,32), activation='relu', solver='adam', max_iter=500))
print(f"Bootstrap Accuracy Scores (ANN): {bootstrap_scores_ann}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(bootstrap_scores, marker='o', linestyle='-')
plt.title('Bootstrap Accuracy Scores (Logistic Regression)')
plt.xlabel('Iteration')
plt.ylabel('Accuracy')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(bootstrap_scores_knn, marker='o', linestyle='-')
plt.title('Bootstrap Accuracy Scores (KNN)')
plt.xlabel('Iteration')
plt.ylabel('Accuracy')
plt.grid(True)
plt.show()
```
